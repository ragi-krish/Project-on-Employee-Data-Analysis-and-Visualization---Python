#### 4. Discover which team and position have the highest salary expenditure

```python
#Discover which team and position have the highest salary expenditure.
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum()
salary_expenditure

max_expenditure = salary_expenditure.idxmax()
max_salary = salary_expenditure.max()
max_expenditure, max_salary

print(f"The team and position with the highest salary expenditure are: Team-->{ max_expenditure[0]},Position-->{max_expenditure[1]}")
print(f"The highest salary expenditure is: ${max_salary}")
```

```python

# Plot bar chart
plt.figure(figsize=(20, 10))
for team in salary_expenditure['Team'].unique():
    team_data = salary_expenditure[salary_expenditure['Team'] == team]
    plt.bar(team_data['Position'], team_data['Salary'], label=team)

# Add labels and legend
plt.title('Salary Expenditure by Team and Position', fontsize=16)
plt.xlabel('Position', fontsize=12)
plt.ylabel('Total Salary ($)', fontsize=12)
plt.legend(title='Team',fontsize=10,loc='upper left', bbox_to_anchor=(1.05, 1))
  # Place outside the plot

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Calculate total salary expenditure for each team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False).head(10)
top_teams = team_salary.index

# Filter data for top 10 teams
filtered_data = df[df['Team'].isin(top_teams)]

# Group data by Team and Position
salary_expenditure = filtered_data.groupby(['Team', 'Position'])['Salary'].sum().unstack(fill_value=0)

# Plot setup
salary_expenditure.plot(kind='bar', stacked=True, figsize=(15, 8), colormap='viridis', alpha=0.85)

# Add labels and customize the chart
plt.title('Top 10 Teams by Salary Expenditure (Stacked by Position)', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Total Salary ($)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Position', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python

# Calculate total salary expenditure for each team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=True).head(5)
last_teams = team_salary.index

# Filter data for last 5 teams
filtered_data = df[df['Team'].isin(last_teams)]

# Group data by Team and Position
salary_expenditure = filtered_data.groupby(['Team', 'Position'])['Salary'].sum().unstack(fill_value=0)

# Plot setup
salary_expenditure.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='coolwarm', alpha=0.85)

# Add labels and customize the chart
plt.title('Last 5 Teams by Salary Expenditure (Stacked by Position)', fontsize=16)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Total Salary ($)', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Position', fontsize=10,bbox_to_anchor=(1.05, 1),loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

#### 4.Investigate if there's any correlation between age and salary, and represent it visually

```python
# Investigate if there's any correlation between age and salary, and represent it visually.
import seaborn as sns

# Ensure the dataset has Age and Salary without missing values
filtered_data = df[['Age', 'Salary']].dropna()

# Calculate the correlation coefficient
correlation = filtered_data.corr().iloc[0, 1]
print(f"Correlation between Age and Salary: {correlation:.2f}")
```

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Salary', data=filtered_data, scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Correlation between Age and Salary', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary ($)', fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Salary'], color='green', alpha=0.6)
plt.title('Scatter Plot of Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```
